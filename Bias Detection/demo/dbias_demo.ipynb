```python
import pandas as pd
```

```python
df = pd.read_csv("ISCA-IUB-dataset.csv")
df.info()
```

```python
df.head(2)
```

```python
# add a new column that is the OR operation of Biased and Calling_Out columns
df['bias'] = (df['Biased'] == 1) | (df['Calling_Out'] == 1)
```

```python
from Dbias.text_debiasing import *;
from Dbias.bias_classification import *;
from Dbias.bias_recognition import *;
from Dbias.bias_masking import *;
import os
import pandas as pd
```

```python
# df = pd.read_csv('sample data.csv')
# df.head()
```

```python
def custom_classification(x):
    classi_out = classify(x)
    return classi_out[0]['label'], classi_out[0]['score']

def custom_recognizer(x):
    biased_words = recognizer(x)
    biased_words_list = []
    for id in range(0, len(biased_words)):
        biased_words_list.append(biased_words[id]['entity'])
    return ", ".join(biased_words_list)

def custom_debiasing(x):
    suggestions = run(x)
    if suggestions == None:
        return ""
    else:
      all_suggestions = []
      for sent in suggestions[0:3]:
        all_suggestions.append(sent['Sentence'])
      return "\n\n".join(all_suggestions)
```

```python
subset_df = df.iloc[:1000]
subset_df.describe()
```

In the subset, 3% content is biased, 32.8% content callout bias.

```python
# running Dbias classfication to classify whether a news article is biased or not
subset_df[['state', 'probability']] = subset_df['Text'].apply(lambda x: custom_classification(x)).to_list()
subset_df
```

```python
# change subset_df bias column to integer
subset_df['bias'] = subset_df['bias'].astype(int)
```

```python
subset_df.describe(include='all')
```

```python
subset_df['prediction'] = subset_df['state'].apply(lambda x: 1 if x == 'Biased' else 0)
```

```python
sum(subset_df['state'] == 1)
```

```python
# Calculate accuracy
accuracy = sum(subset_df['bias'] == subset_df['prediction']) / len(subset_df['bias'])

# Calculate true positives, false positives, and false negatives
tp = sum((subset_df['bias'] == 1) & (subset_df['prediction'] == 1))
fp = sum((subset_df['bias'] == 0) & (subset_df['prediction'] == 1))
fn = sum((subset_df['bias'] == 1) & (subset_df['prediction'] == 0))

# Calculate precision, recall, and F1 score
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)

print(f"Accuracy: {accuracy}")
print(f"F1 Score: {f1}")
print(f"Recall: {recall}")
```
